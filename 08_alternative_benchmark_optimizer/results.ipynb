```python
from glob import glob
import json
import pandas as pd
```

```python
files = glob("results/*.json")
```

```python
with open(files[0], "r") as f:
    content = json.loads(f.read())
```

```python
content
```

```python
l_classifier = []
l_mse = [] = []
l_linke = []
l_gruene = []
l_spd = []
l_fdp = []
l_cdu = []
l_afd = []
for file in files:
    with open(file, "r") as f:
        content = json.loads(f.read())
        l_classifier.append(content.get('classifier'))
        l_mse.append(content.get('unweighted_circ_mse'))
        l_linke.append(content.get('optimized_angles').get('linke'))
        l_gruene.append(content.get('optimized_angles').get('gruene'))
        l_spd.append(content.get('optimized_angles').get('spd'))
        l_fdp.append(content.get('optimized_angles').get('fdp'))
        l_cdu.append(content.get('optimized_angles').get('cdu'))
        l_afd.append(content.get('optimized_angles').get('afd'))
```

```python
df = pd.DataFrame({
    'classifier':l_classifier,
    'mse':l_mse,
    'linke':l_linke,
    'gruene':l_gruene,
    'spd':l_spd,
    'fdp':l_fdp,
    'cdu':l_cdu,
    'afd':l_afd,
})
```

```python
df = df.sort_values(by=['mse']).reset_index(drop=True)
```

```python
df.to_json("results.json")
```
